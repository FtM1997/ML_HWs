```python
# Fatemeh Makhloughi
# Machine Learning
# HW10 - KMeans for Mall_Customers/ clustering problem
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score
```

```python
df = pd.read_csv('/content/Mall_Customers.csv')
df.head()
```

```python
x = df.iloc[:,3:]
```

```python
K=range(2,12)
wss = []
for k in K:
  kmeans=KMeans(n_clusters=k)
  kmeans=kmeans.fit(x)
  wss_iter = kmeans.inertia_
  wss.append(wss_iter)
```

```python
plt.xlabel('K')
plt.ylabel('inertia (WSS Error)')
plt.plot(K,wss)
```

```python
kmeans = KMeans(n_clusters=5)
kmeans.fit(x)
```

```python
y_kmeans = kmeans.predict(x)
```

```python
plt.scatter(x.iloc[:, 0], x.iloc[:, 1], c=y_kmeans, s=50)
#c: color, s: marker size
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.5);
```

```python
x = df.iloc[:,[2,4]]
```

```python
Sil = []
K=range(2,12)
for k in K:
  kmeans=KMeans(n_clusters=k)
  kmeans=kmeans.fit(x)
  Sil_iter = silhouette_score(x, kmeans.labels_)
  Sil.append(Sil_iter)
```

```python
plt.xlabel('K')
plt.ylabel('Silhouette score')
plt.plot(K,Sil)
```

```python
kmeans = KMeans(n_clusters=4)
kmeans.fit(x)
```

```python
y_kmeans = kmeans.predict(x)
```
