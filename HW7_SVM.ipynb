```python
# Fatemeh Makhloughi
# Machine Learning
# HW7 - SVM classifier for digits dataset
from sklearn.datasets import load_digits
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
import numpy as np
```

```python
digits = load_digits()
```

```python
x = digits.data
y = digits.target
```

```python
type(x)
```

```python
np.shape(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
svc_classifier = SVC(kernel='linear', C=1)
svc_classifier.fit(x_train, y_train)
```

```python
print("Accuracy: %0.2f" % svc_classifier.score(x_test, y_test))
```

```python
score = cross_val_score(svc_classifier, x, y, cv = 5)
```

```python
print("Mean accuracy: %0.2f with a standard deviation of %0.2f" % (score.mean(),score.std()))
```

```python
grid_parameters = {'C': [1,5, 10, 50, 100], 'gamma':[0.001, 0.01, 0.1, 1, 10],'kernel': ['rbf','linear']}
grid = GridSearchCV(SVC(),grid_parameters)
grid.fit(x_train,y_train)
print("Grid Best Result:",grid.best_params_)
```

```python
pred = grid.predict(x_test)
```

```python
print("accuracy =  %0.2f" % grid.score(x_test, y_test))
```

```python
print(confusion_matrix(y_test, pred))
```
