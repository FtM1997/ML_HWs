```python
# Fatemeh Makhloughi
# Machine Learning
# HW4- Linear Regression for Insurance Dataset (Regression problem)
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv('/content/drive/MyDrive/Machine Learning/insurance.csv')
```

```python
df
```

```python
X = df.iloc[:,:-1]
Y = df['charges']
```

```python
#label_encoder_x= LabelEncoder()
cols = ['gender', 'smoker', 'region']
X[cols] = X[cols].apply(LabelEncoder().fit_transform)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
Scaler = MinMaxScaler()
x_train = Scaler.fit_transform(x_train)
x_test = Scaler.transform(x_test)
```

```python
lin_reg_model = LinearRegression()
lin_reg_model.fit(x_train, y_train)
```

```python
y_pred = lin_reg_model.predict(x_test)
```

```python
plt.scatter(y_test, y_pred)
plt.xlabel('Charges')
plt.ylabel('predicted Charges')
plt.show()
```

```python
r2 = r2_score(y_test,y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)
print("R2_Score =",r2)
print("RMSE =",rmse)
```

Linear regression is not suitable for this problem because RMSE is high.
